```python
# 1. Analisi Esplorativa (EDA) per comprendere il dataset
print("\nPrime 10 righe del dataset:")
print(df.head(10))
# Verifica dei valori unici nella colonna 'location'
print("Valori unici in 'location':")
print(df['location'].unique())
```

```python
# Il dataset location contiene dati aggregati perché se noti che compaiono valori come "World" oppure "Europe" oppure "European Union (27)" per esempio
# Verifica dei valori unici nella colonna 'continent'
print("\nValori unici in 'continent':")
print(df['continent'].unique())
```

```python
df_aggregati = df[df['location'] == 'Europe']
print("\nRighe aggregate:")
df_aggregati.head()
```

```python
# Per le analisi a livello di continente, filtreremo le righe dove la colonna continent non è nulla.
# Visualizza le prime righe per comprendere i valori
print("\nEsempi di 'new_cases':")
df[['date', 'location', 'new_cases']].head(10)
```

```python
print("\nEsempi di 'total_cases':")
df[['date', 'location', 'total_cases']].head(10)
```

```python
# 2. Analisi per continente
# Filtrare i dati a livello di paese: consideriamo solo le righe dove 'continent' non è NaN
df_country = df[df['continent'].notna()].copy()
# Per ogni paese, selezioniamo l'ultima registrazione (ordinando per data)
df_last = df_country.sort_values('date').groupby('location').last().reset_index()
# Raggruppare per continente e sommare i total_cases dei singoli paesi
continent_cases = df_last.groupby('continent')['total_cases'].sum().reset_index()
# Calcolare il totale mondiale dei casi
world_total = df_last['total_cases'].sum()
# Calcolare la percentuale dei casi per ciascun continente
continent_cases['percentage'] = continent_cases['total_cases'] / world_total * 100
print("Casi totali per continente e percentuale rispetto al totale mondiale:")
print(continent_cases)
```

```python
# 3. Analisi specifica per l’Italia nel 2022
italy_2022 = df[(df['location'] == 'Italy') & 
                (df['date'] >= '2022-01-01') & 
                (df['date'] <= '2022-12-31')].copy()
# Convertire la colonna 'date' in datetime
italy_2022['date'] = pd.to_datetime(italy_2022['date'])
# Filtrare via le righe senza misurazioni dei nuovi casi
italy_2022 = italy_2022[italy_2022['new_cases'].notna()]
# Grafico: evoluzione dei casi totali
plt.figure(figsize=(10,5))
plt.plot(italy_2022['date'], italy_2022['total_cases'], marker='o')
plt.title("Evoluzione dei casi totali in Italia nel 2022")
plt.xlabel("Data")
plt.ylabel("Casi totali")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Grafico: andamento dei nuovi casi
plt.figure(figsize=(10,5))
sns.lineplot(x='date', y='new_cases', data=italy_2022, color='orange', marker='o')
plt.title("Nuovi casi in Italia nel 2022")
plt.xlabel("Data")
plt.ylabel("Nuovi casi")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4. Boxplot dei pazienti in terapia intensiva (ICU) per Italia, Germania e Francia
# Dati per Italia, Germania e Francia (Maggio 2022 - Aprile 2023)
countries = ['Italy', 'Germany', 'France']
mask = (df['location'].isin(countries)) & (pd.to_datetime(df['date']) >= '2022-05-01') & (pd.to_datetime(df['date']) <= '2023-04-30')
df_box = df[mask].copy()
df_box['date'] = pd.to_datetime(df_box['date'])

# Creazione del boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='location', y='icu_patients', data=df_box)
plt.title("Distribuzione dei pazienti in ICU (Maggio 2022 - Aprile 2023)")
plt.xlabel("Nazione")
plt.ylabel("Pazienti in terapia intensiva (ICU)")
plt.show()
```

```python
# Se può vedere che tra maggio 2022 e Aprile 2023 c'era maggiore pressione sul sistema sanitario nella Germania perché hanni più pazienti in terapia intensiva a differenza dell’Italia, che aveva meno pazienti in terapia intensiva.

# 5. Analisi dei pazienti ospedalizzati per Italia, Germania, Francia e Spagna nel 2021
# Dati per Italia, Germania, Francia e Spagna nel 2021
countries2 = ['Italy', 'Germany', 'France', 'Spain']
mask2 = (df['location'].isin(countries2)) & (pd.to_datetime(df['date']) >= '2021-01-01') & (pd.to_datetime(df['date']) <= '2021-12-31')
df_2021 = df[mask2].copy()
df_2021['date'] = pd.to_datetime(df_2021['date'])

# Calcolare la somma dei pazienti ospedalizzati per ogni nazione
sum_hosp = df_2021.groupby('location')['hosp_patients'].sum().reset_index()
print("Somma dei pazienti ospedalizzati nel 2021 per ciascuna nazione:")
print(sum_hosp)
```

```python
# Visualizzazione grafica con bar plot
plt.close('all')
plt.figure(figsize=(8,6))
sns.barplot(x='location', y='hosp_patients', data=sum_hosp, hue='location', palette="viridis", dodge=False, legend=False)
plt.title("Somma dei pazienti ospedalizzati nel 2021")
plt.xlabel("Nazione")
plt.ylabel("Pazienti ospedalizzati (somma)")
plt.show()
```

```python
# Verifica dei dati nulli nella colonna 'hosp_patients'
nulls = df_2021['hosp_patients'].isna().sum()
print("Numero di valori nulli in 'hosp_patients' nel 2021:", nulls)
```

```python
# Numero totale di righe nel dataset per il 2021 (nazioni selezionate)
total_rows = len(df_2021)
nulls = df_2021['hosp_patients'].isna().sum()
percentage_nulls = (nulls / total_rows) * 100

print("Numero di valori nulli in 'hosp_patients' nel 2021:", nulls)
print("Percentuale di dati nulli: {:.2f}%".format(percentage_nulls))
```
